```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Dataset .csv")
data.head()
```

```python
data.columns
```

```python
plt.hist(data["Aggregate rating"], bins=10)
plt.title("Distribution of Restuarant Ratings")
plt.xlabel("Aggregate Ratings")
plt.ylabel("Number of Restuarants")
plt.show()
```

```python
rating_bins=pd.cut(data["Aggregate rating"],bins=[0,1,2,3,4,5])
rating_distribution=rating_bins.value_counts().sort_index()
rating_distribution
```

```python
average_votes = data["Votes"].mean()
average_votes
```

Conclusion:
The distribution of restaurant ratings shows that the most common rating range is 3 to 4, with the highest number of restaurants falling within this category. This indicates that a majority of restaurants receive moderately high ratings, reflecting generally positive customer experiences.
Additionally, the average number of ratings (votes) received by restaurants is approximately 157, suggesting a reasonable level of customer engagement across the dataset. Overall, the data indicates that most restaurants are fairly well-rated and actively reviewed by customers.
