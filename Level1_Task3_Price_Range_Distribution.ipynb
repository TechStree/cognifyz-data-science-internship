```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv("Dataset .csv")
data.head()
```

```python
data.columns
```

```python
price_counts = data["Price range"].value_counts().sort_index()
price_counts
```

```python
total_resturants=data.shape[0]
price_percentage=(total_resturants/price_counts)*100
price_percentage
```

```python
price_counts.plot(kind="bar")
plt.title("Distribution of Restaurant Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

Conclusion:
The price range distribution shows that the majority of restaurants fall under Price Range 1, followed by Price Range 2, indicating that most restaurants offer affordable to moderately priced dining options.
Higher price ranges (Price Range 3 and 4) have significantly fewer restaurants, suggesting that premium and luxury dining options are less common in the dataset.
