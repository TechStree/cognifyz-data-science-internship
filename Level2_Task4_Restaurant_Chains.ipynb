```python
import pandas as pd
```

```python
data = pd.read_csv("Dataset .csv")
data.head()
```

```python
restaurant_counts = data["Restaurant Name"].value_counts()
chains = restaurant_counts[restaurant_counts > 1]
chains.head(10)
```

```python
chains_df = data[data["Restaurant Name"].isin(chains.index)]

chain_popularity = chains_df["Restaurant Name"].value_counts().head(10)
chain_popularity
```

```python
chain_ratings = (
    chains_df
    .groupby("Restaurant Name")["Aggregate rating"]
    .mean()
    .sort_values(ascending=False)
)

chain_ratings.head(10)
```

```python
top_chains = chain_popularity.index
top_chain_ratings = chain_ratings.loc[top_chains]
top_chain_ratings
```

Conclusion:
The analysis confirms the presence of several restaurant chains in the dataset, identified by multiple outlets operating under the same restaurant name. Popular chains such as Cafe Coffee Day, Domino’s Pizza, and Subway have the highest number of outlets, indicating strong brand presence and wide market reach.

When analyzing ratings, some highly popular chains show moderate average ratings. For example, Cafe Coffee Day and Domino’s Pizza have average ratings in the range of approximately 2.4–2.7, suggesting that popularity does not always correspond to higher customer satisfaction.

On the other hand, certain chains such as Barbeque Nation demonstrate both a strong presence and a high average rating (above 4.3), indicating a balance between popularity and quality. Additionally, chains like AB’s Absolute Barbecues, Naturals Ice Cream, and The Cheesecake Factory receive very high average ratings despite having fewer outlets.

Overall, the analysis highlights that while some restaurant chains are widely popular, higher customer ratings are often associated with select premium or niche chains rather than mass-market brands.
