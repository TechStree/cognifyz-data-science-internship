```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Dataset .csv")
data.head()
```

```python
data.columns
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data["Longitude"], data["Latitude"], alpha=0.5)
plt.title("Geographic Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
geo_data = data[(data["Latitude"] != 0) & (data["Longitude"] != 0)]

plt.figure(figsize=(8,6))
plt.scatter(geo_data["Longitude"], geo_data["Latitude"], alpha=0.5)
plt.title("Restaurant Clusters by Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

Conclusion:
The geographic visualization of restaurant locations using latitude and longitude reveals clear clustering patterns across specific regions. A high concentration of restaurants can be observed in certain areas, indicating that restaurants are predominantly located in urban and metropolitan regions.

These clusters suggest strong restaurant density in major cities and economically active areas, while other regions show sparse or scattered restaurant locations. This uneven distribution highlights how restaurant presence is influenced by factors such as population density, urbanization, and market demand.

Overall, the analysis confirms that restaurants tend to cluster in well-developed regions rather than being uniformly distributed across all geographic locations.
