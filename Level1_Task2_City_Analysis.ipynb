```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Dataset .csv")
data.head()
```

```python
data.columns
```

```python
city_counts=data["City"].value_counts()
city_counts.head()
```

```python
top_city=city_counts.idxmax()
top_city
```

```python
city_counts.max()
```

```python
city_avg_rating=data.groupby("City")["Aggregate rating"].mean()
city_avg_rating.head()
```

```python
best_rated_city=city_avg_rating.idxmax()
best_rated_city
```

```python
city_avg_rating.max()
```

```python
city_counts.head(10).plot(kind="bar")
plt.title("Top 10 cities by number of resturants")
plt.xlabel("City")
plt.ylabel("Number of restuarants")
plt.show()
```

Conclusion:
The analysis reveals that New Delhi has the highest number of restaurants in the dataset, indicating a dense and competitive food market in the city.
Additionally, Inner City has the highest average restaurant rating of 4.9, suggesting a higher level of customer satisfaction compared to other cities.
