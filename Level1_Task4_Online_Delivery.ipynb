```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Dataset .csv")
data.head()
```

```python
data.columns
```

```python
delivery_counts = data["Has Online delivery"].value_counts()
delivery_counts
```

```python
total_restaurants = data.shape[0]

delivery_percentage = (delivery_counts / total_restaurants) * 100
delivery_percentage
```

```python
avg_rating_delivery = data.groupby("Has Online delivery")["Aggregate rating"].mean()
avg_rating_delivery
```

```python
avg_rating_delivery.plot(kind="bar")
plt.title("Average Rating: Online Delivery vs No Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")
plt.show()
```

Conclusion:
The analysis reveals that 25.66% of restaurants offer online delivery, while 74.34% do not provide online delivery services.
When comparing average ratings, restaurants with online delivery have a higher average rating of approximately 3.25, whereas restaurants without online delivery have a lower average rating of about 2.47.
This indicates that restaurants offering online delivery tend to receive better customer ratings, suggesting a positive impact of online delivery services on customer satisfaction.
