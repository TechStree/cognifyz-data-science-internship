```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv("Dataset .csv")
data.head()
```

```python
data.info()
```

```python
cuisine_column=data["Cuisines"]
cuisine_column.head()
```

```python
cuisine_clean=cuisine_column.dropna()
```

```python
cuisine_split=cuisine_clean.str.split(", ")
```

```python
cuisine_exploded=cuisine_split.explode()
```

```python
top_cuisines=cuisine_exploded.value_counts().head(3)
top_cuisines
```

```python
total_restuarants=data.shape[0]
cuisine_percentage=(top_cuisines/total_restuarants)*100
cuisine_percentage
```

```python
top_cuisines.plot(kind="bar")
plt.title("Top 3 common cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Number of resturants")
plt.show()
```

Conclusion:
The analysis shows that the most common cuisines in the dataset are North Indian, Chinese, and Fast Food.

These cuisines are served by a significant percentage of restaurants, indicating their popularity among customers.
